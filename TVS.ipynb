```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pathlib import Path
```

```python
team_stats_path = "teamstats2012-2021.csv"
```

```python
team_stats=pd.read_csv(team_stats_path)
```

```python
df1=team_stats
df1.head(5)
```

```python
df1_2=df1.query("year == 2021")
```

```python
df1_2
```

```python
df2=df1_2.groupby(['team'])['Rebounds','Inside 50s','Clearances','Contested Possessions'].sum()
```

```python
df2
```

```python
df3 = df2.sum(axis = 1)
```

```python
df4=pd.DataFrame(df3)
```

```python
df4.columns=['TVS']
```

```python
TVS=df4.sort_values(['TVS'],ascending=False)
```

```python
TVS
```

```python
filepath = Path('TVS.csv')
TVS.to_csv(filepath)
```
